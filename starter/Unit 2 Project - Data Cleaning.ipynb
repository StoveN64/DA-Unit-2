### Step 2: Treat Names / Casing / Spelling Issues

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###

# Renaming all columns, changing column names to lowercase and snake_case, removing punctuation, and dropping 'Unnamed'.

### CLEAN ### 

def snake_case(dataframe):
    dataframe.columns = dataframe.columns.str.lower()
    dataframe.columns = dataframe.columns.str.replace(r'[^\w\s]+', '')
    dataframe.columns = dataframe.columns.str.replace(' ', '_')
    return dataframe

df1 = snake_case(df1)
df2 = snake_case(df2)

df1.drop(df1.columns[[0]], axis=1, inplace=True)
df2.drop(df2.columns[[0]], axis=1, inplace=True)

### TEST ###
df1.head()
df2.head()
```

### Step 3: Treat Missing Values

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
df1.info()
df2.info()
```

```python
### DEFINE ###


### CLEAN ### 

df1.bathrooms = df1.bathrooms.fillna(df1.bathrooms.mean())
df1.sqft_lot = df1.sqft_lot.fillna(df1.sqft_lot.mean())
df1.waterfront = df1.waterfront.fillna("No")
view_mode = df1.view.mode()
df1.view = df1.view.fillna(value=view_mode[0])
house_condition_mode = df1.house_condition.mode()
df1.house_condition = df1.house_condition.fillna(house_condition_mode[0])
grade_mode = df1.grade.mode()
df1.grade = df1.grade.fillna(grade_mode[0])
df1.sqft_above = df1.sqft_above.fillna(df1.sqft_above.mean())
df1.dropna(subset=['year_built'], inplace=True)
df1.drop(columns=['year_renovated'], inplace=True)
df1.dropna(subset=['lat', 'long'], inplace=True)
df2.dropna(subset=['price'], inplace=True)
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)

### TEST ###
df1.isnull().sum()
df2.isnull().sum()
df2.head()
df2.head()
```

### Step 4: Treat Outliers

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
from scipy.stats import zscore
```

```python
outliers_arr = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'grade', 'sqft_above', 'sqft_basement', 'year_built', 'sqft_living_15', 'sqft_lot_15']

for i in outliers_arr:
    df1[f'{i}_zscore'] = zscore(df1[f'{i}'])



df1.head()
```

```python
### DEFINE ###

# Year renovated to be sorted out via z-score

### CLEAN ### 

for i in outliers_arr:
    if i == 'year_renovated': # not relevant to sort out.
        df1.drop(df1[(df1[f'{i}_zscore'] > 4) | (df1[f'{i}_zscore'] < -4)].index, inplace=True)
    else:
        df1.drop(df1[(df1[f'{i}_zscore'] > 3) | (df1[f'{i}_zscore'] < -3)].index, inplace=True)

### TEST ###

# df1.info()
```

### Step 2: Choose Method to Re-Express Variables

### Step 3: Perform Transformation 

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
waterfront_dictionary = {
    'No': 0,
    'Yes': 1
}

view_dictionary = {
    'Very Poor': 1,
    'Poor': 2,
    'Neutral': 3,
    'Good': 4,
    'Great': 5
}

condition_dictionary = {
    'Very Bad Condition': 1,
    'Bad Condition': 2,
    'OK Condition': 3,
    'Good Condition': 4,
    'Great Condition': 5
}

df1.waterfront = df1.waterfront.map(waterfront_dictionary).astype('int64')
df1.view = df1.view.map(view_dictionary).astype('int64')
df1.house_condition = df1.house_condition.map(condition_dictionary).astype('int64')

df1.info()
```
